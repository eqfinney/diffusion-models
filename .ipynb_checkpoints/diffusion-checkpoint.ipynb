```python
import matplotlib.cm as cm
```

```python
height, width = 25,25
```

```python
TWOPI = np.pi * 2
```

```python
img = np.random.uniform(0,1,height*width)
```

```python
imgplot = plt.imshow(img.reshape(height,width), cm.Greys_r, interpolation='none')
```

```python
def angleToVec(theta):
    return np.array([np.cos(theta*TWOPI), np.sin(theta*TWOPI)])
```

```python
angleToVec = np.frompyfunc(angleToVec, 1, 1)
```

```python
gradients = angleToVec(img).reshape(height,width)
```

```python
def dotGridGradient(i, x):
    ix = i[0]
    iy = i[1]
    g = gradients[ix][iy]
    return np.dot(g, x-i)
```

```python
def scurve(x, upper_left):
    p = x - upper_left
    s = 3*(p**2) - 2*(p**3)
    return s
```

```python
def better_according_to_perlin(x, x0):
    p = x - x0
    s = 6*(p**5) - 15*p**4 + 10*(p**3)
    return s
```

```python
def noise(x, avg_func=scurve):
    # get grid coordinates
    x0 = upper_left = np.floor(x)
    x1 = lower_right = upper_left + 1
    upper_right = np.array([x1[0], x0[1]])
    lower_left = np.array([x0[0], x1[1]])
    
    grid_points = np.array([upper_left, upper_right, lower_left, lower_right])
    
    weights = np.apply_along_axis(dotGridGradient, 1, grid_points, x)
    
    # s-curve
    s = avg_func(x, upper_left)
    
    def lerp(start, end, t):
        return (1.0 - t)*start + t*end
    
    val = lerp(lerp(weights[0], weights[1], s[0]),
               lerp(weights[2], weights[3], s[0]),
               s[1],
              )
    
    return val
    
```

```python
#noise_vectorized = np.vectorize(noise)
```

```python
def cartesian_product(x, y):
    return np.transpose([np.tile(x, len(y)), np.repeat(y, len(x))])
```

```python
dots_per_square = 20
```

```python
xs = np.linspace(0,height-1, endpoint=False, num=width*dots_per_square)
```

```python
points = cartesian_product(xs, xs)
```

```python
noise_img = np.apply_along_axis(noise, 1, points, avg_func=scurve)
```

```python
noiseplot = plt.imshow(noise_img.reshape(width*dots_per_square,width*dots_per_square), cm.Greys_r, interpolation='none')
```

```python
#noise_img = np.apply_along_axis(noise, 1, points, avg_func=lambda x, y: [0.5,0.5])
```

```python
img = noise_img.reshape(width*dots_per_square,width*dots_per_square)
```

```python
import png
```

```python
def write_to_png(filename, img):
    f = open(filename, 'wb')
    w = png.Writer(len(img),len(img[0]), greyscale=True)
    w.write(f, ((img-img.min())*(255.0/(img.max()-img.min()))).astype(int))
    f.close();
```

```python
write_to_png('/tmp/perlin_noise_3rd_degree.png', img)
```
